```python
import pandas as pd
import pymongo
```

```python
client = pymongo.MongoClient()
database = client['exercices']
collection = database['youtube']
```

```python
df_youtube = pd.read_csv("./data/USvideos.csv")
df_youtube.head(100)
```

```python
df_youtube['channel_title'].unique()
df_youtube['category_id'].unique()
```

## Question 0

### Netoyer les données

### Importer les données

```python
documents = df_youtube.to_dict(orient = 'records')
```

```python
collection.insert_many(documents)
```

## Question 1  

```python
#Récupérer toutes les vidéos de la chaîne Apple.
video_apple = collection.find({"channel_title": "Apple"})
list(video_apple)
```

## Question 2

```python
#Compter le nombre de catégories différentes
nb_categ = collection.aggregate([{"$group" : {"_id" : "$category_id"}}])
len(list(nb_categ))
```

## Question 3

```python
#Découper les tags en listes et mettre à jour les tags de chacun des documents avec une requête update

tags = collection.find({},{"tags":1,"_id":1})
liste = [(x["tags"].split('|'),x["_id"]) for x in list(tags)]
for tag, _id in liste:
    collection.update_one( {"_id":_id}, {"$set":{"tags":tag}} )
collection.find_one()
```

## Question 4

```python
#Récupérer les vidéos les plus vues.
videos_vues = collection.find().sort([("views", -1)]).limit(10)
list(videos_vues)
```

## Question 5

```python
#Compter le nombre moyen de vues en fonction de la catégorie.
vue_categ = collection.aggregate([{"$group" : {"_id" : "$category_id", "moyenne" : {"$avg" : '$views'}}}])
list(vue_categ)
```

## Question 6 

```python
#Récupérer les chaines Youtube avec la plus grande moyenne de likes.
chaine_like = collection.aggregate([{"$group" : {"_id" : "$channel_title", "moyenne" : {"$avg" : '$likes'}}},
                                    {"$sort" : {'moyenne' : -1}},
                                    {"$limit" : 10}])
list(chaine_like)
```
